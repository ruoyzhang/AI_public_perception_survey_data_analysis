```python
import pandas as pd
import pickle
from matplotlib import pyplot as plt
import os
```

```python
data_dir_CN = os.path.join('..','..','AI_perception_survey_data', 'CN', 'cleaned')
```

```python
# Reading data
# CN
with open(os.path.join(data_dir_CN, 'wc_all.pickle'), 'rb') as handle:
    data_cn = pickle.load(handle)
```

```python
data_cn.head()
```

we forgot to convert certain questions into the positive 1-7 likert scale

```python
# Questions to modify
likert_q_num = [4,5,8,9,12,13]
likert_q_num = ['q{:02}'.format(q_num) for q_num in likert_q_num]
```

```python
# checking dtypes
data_cn.dtypes
```

So the data types are good, let's rescale

```python
for q_num in likert_q_num:
    data_cn[q_num] = data_cn[q_num] + 4
```

```python
data_cn.head()
```

```python
# resave
with open(os.path.join(data_dir_CN, 'wc_all.pickle'), 'wb') as handle:
    pickle.dump(data_cn, handle)
```
